```python
# Quartiles

# Q1 - 25
# Q2 - 50
# Q3 -75
```

```python
q1=data['Population in thousands (2017)'].quantile(0.25)
q1
```

```python
q2=data['Population in thousands (2017)'].quantile(0.50)
q2
```

```python
q3=data['Population in thousands (2017)'].quantile(0.75)
q3
```

```python
# Inter Quartile Range (IQR=Q3-Q1)

iqr=q3-q1
iqr
```

```python
# Upper fence : Q3 + 1.5*IQR

uf=q3 + (1.5*iqr)
uf
```

```python
# Lower fence : Q1 - 1.5*IQR

lf=q1 - (1.5*iqr)
lf
```

```python
# ANY VALUE LESS THAN LOWER FENCE i.e. -27712.0 & GREATER THAN UPPER FENCE i.e 47336.0 ARE OUTLIERS
```

```python
# only those record which are outliers will be displayed
data[(data['Population in thousands (2017)']<lf) | (data['Population in thousands (2017)']>uf)]
```
